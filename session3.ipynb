# inheritance

- Accessing properties and functions from one class to another class is called as inheritence

# single inheritance

```python
# parent class / Base class / super class
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def print_message(self):
        print(f"Name={self.name}\nAge={self.age}")
```

```python
# child class / Derive class / subclass
class Employee(Person):
    def __init__(self,name,age,department):
        super().__init__(name,age)
        self.department=department
    def display(self):
        print(f"{self.print_message()}\nDepartment={self.department}")
```

```python
e1 = Employee("Sakshi",21,"Data Analytics")
```

```python
e1.display()
```

# Multiple inheritance

```python
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def show(self):
        print(f"hello my name is {self.name} and my salary is {self.salary}")
```

```python
class Teamleader(Employee):
    def __init__(self,name,salary,department):
        super().__init__(name,salary)
        self.department=department
    def display(self):
        print(f"{self.show()}.I am team leader of {self.department} department")
```

```python
class Manager(Teamleader):
    def __init__(self,name,salary,department):
        super().__init__(name,salary,department)
    def display_info(self):
        print(f"{self.display()}. I am manager of {self.department} department")
```

```python
#object
m1=Manager("Akhil",45000,"Data Science")
```

```python
m1.display_info()
```

# hierarchical

```python
class Vehicle:
    def __init__(self,name,color):
        self.color=color
        self.name=name
    def print_data(self):
        print(f"Name={self.name}\ncolor={self.color}")
```

```python
class Twowheeler(Vehicle):
    def __init__(self,name,color,no_of_wheels):
        super().__init__(name,color)
        self.no_of_wheels=no_of_wheels
    def print_message(self):
        print(f"{self.print_data()}\nNumber of wheels={self.no_of_wheels}")
```

```python
class Fourwheeler(Vehicle):
    def __init__(self,name,color,seating_capacity):
        super().__init__(name,color)
        self.seating_capacity=seating_capacity
    def display(self):
        print(f"{self.print_data()}\n seating capacity={self.seating_capacity}")
```

```python
t1=Twowheeler("Honda","black",2)
```

```python
t1.print_message()
```

```python
f1=Fourwheeler("BMW" , "White" , 5)
f1.display()
```
