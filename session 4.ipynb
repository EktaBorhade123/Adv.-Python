# function overriding

```python
class India :
    def capital(self):
        print("the capital of India is new Delhi")
    def language(self):
        print("the official language of India is Hindi")
```

```python
class USA :
    def capital(self):
        print("Capital of USA is Washington DC")
    def language(self):
        print("official language of USA is English ")
```

```python
I=India()
U=USA()
```

```python
I.capital()
I.language()
U.capital()
U.language()
```

```python
for i in I , U:
    i.capital()
    i.language()
```

```python
class cal_area:
    def __init__(self,a,b=0):
        self.a=a
        self.b=b
        self.aos=a*a
        self.aor=a*b
    def area_of_sqaure(self):
        print("area of square=",self.aos)
    def area_of_rectangle(self):
        print("area of rectangle=",self.aor)
```

```python
a1=cal_area(5)
```

```python
a1.area_of_sqaure()
```

```python
a1.area_of_rectangle()
```

```python
a2=cal_area(8,9)
```

```python
a2.area_of_sqaure()
```

```python
a2.area_of_rectangle()
```

# operator overloading 


```python
# + = __add__
# - = __sub__
# / =__div__
# < = __lt__ 
# > = __gt__ 
# & = __and__ 
# | = __or__
```

```python
class Add :
    def __init__(self,a):
        self.a=a
    def __add__(self,other):
        return self.a + other.a
```

```python
a1=Add(5)
a2=Add(10)
```

```python
a1+a2
```

```python
a3=Add(20)
a4=Add(50)
```

```python
a3+a4
```

```python
class compare:
    def __init__(self,a):
        self.a=a
    def __gt__(self,other):
        if self.a>other.a:
            print(f"my first object is greater than second object")
        else:
            print("my second object is greater than first object")
```

```python
c1=compare(10)
```

```python
c2=compare(20)
```

```python
c2>c1
```

```python
class complex:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __add__(self,other):
        return f"{self.a+ other.a} + i{self.b + other.b}"
```

```python
n1 = complex(10,15)
n2 = complex(20,25)
```
