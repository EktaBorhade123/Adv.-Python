```python
lst = [2,3,4,5,6,7,8,12,15,18]
```

```python
def cal_sum(l):
    sum=0
    for i in l:
        sum=sum+i
        print("Total_sum=",sum)
```

```python
cal_sum(lst)
```

```python
def cal_sq(l):
    for i in l:
        print(i*i)
```

```python
cal_sq(lst)
```

```python
def cal_cube(l):
    for i in l:
        print(i*i*i)
```

```python
cal_cube(lst)
```

```python
# hypotenius of right angled triangle
```

```python
def hypotenius_tri():
    a=int(input("Enter 1st number"))
    b=int(input("enter 2nd number"))
    c=((a*a)+(b*b))**0.5  
    print(f"hypotenius of right angled triable is {c}")
```

```python
hypotenius_tri()
```

```python
# wapp which takes 3 sides and determines whether 3 sides are forming a trianle or not
```

```python
def cal_tri(a,b,c):
    if ((a+b)>c) and ((a+c)>b) and ((b+c)>a):
        print("these 3 sides are forming a triable")
    else:
        print("these 3 sides are not forming a triagle")
```

```python
cal_tri(1,2,3)
```

```python
cal_tri(12,14,18)
```
