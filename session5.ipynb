# try except error handling

```python
try:
    a=int(input("Enter a number"))
    b = int(input("enter 2nd number"))
    c=a/b
    print(c)
except ZeroDivisionError:
    print("you are trying to divide a number by zero")
```

```python
try:
    lst=[2,5,6,8,9,10,23,34]
    idx=int(input("enter a number"))
    print(lst[idx])
except IndexError as e:
    print(e)
```

```python
try:
    a='23'
    b=23
    c=a+b
    print(c)
except TypeError as e:
    print(e)
```

```python
try:
    age=int(input("enter your age"))
    if age<0 or age>100:
        raise ValueError("age ust be between 0 to 100")
    print("age is varified")
except ValueError as e:
    print(e)
```

# finally

```python
try:
    age=int(input("enter your age"))
    if age<0 or age>100:
        raise ValueError("age ust be between 0 to 100")
    print("age is proper")
except ValueError as e:
    print(e)
finally:
    print("finally age is varified")
```
